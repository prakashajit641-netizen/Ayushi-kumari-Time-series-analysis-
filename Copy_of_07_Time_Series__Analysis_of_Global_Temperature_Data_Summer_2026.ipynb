```python
import pandas as pd
import numpy as  np

# Load the downloaded CSV file
df = pd.read_csv('monthly_csv.csv')

# Display the first 5 rows
df.head()
```

```python
# Convert 'Date' column from string to datetime object
df['Date'] = pd.to_datetime(df['Date'])

# Sort the entire DataFrame by 'Date' in ascending order
df = df.sort_values(by='Date', ascending=True)

# Display the first 5 rows of the sorted DataFrame
df.head()
```

### Create 'Year' and 'Month' columns and filter data

```python
# Print the shape of the DataFrame (rows, columns)
print("DataFrame Shape:", df.shape)

# Check for the total number of missing (null) values per column
print("\nMissing values per column:")
print(df.isnull().sum())

# To print the absolute grand total of missing values across the whole DataFrame:
# print("\nTotal missing values in entire DataFrame:", df.isnull().sum().sum())
```

```python
# Ensure 'Date' column is datetime type (already done in previous steps)
df['Date'] = pd.to_datetime(df['Date'])

# Extract Year and Month
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

# Display the head of the DataFrame with the new columns
df.head()
```

```python
# Ensure the data is sorted chronologically before rolling calculations
df = df.sort_values(by=['Source', 'Date']).reset_index(drop=True)

# Calculate 12-month rolling mean of 'Mean' temperature grouped by 'Source'
df['Moving_Avg'] = df.groupby('Source')['Mean'].transform(lambda x: x.rolling(window=12).mean())

# Display the last 5 rows
df.tail()
```

```python
# Find the latest year in the dataset
latest_year = df['Year'].max()

# Filter for the last 20 years (inclusive of the latest year)
df_last20 = df[df['Year'] >= (latest_year - 19)]

# Print the shape of the new DataFrame
print(df_last20.shape)
```

```python
# Group by 'Source' and calculate the mean of the 'Mean' column
source_means = df.groupby('Source')['Mean'].mean()

# Display the resulting Pandas Series
print(source_means)
```

```python
# Group by 'Year' and calculate the mean of the 'Mean' column
yearly_means = df.groupby('Year')['Mean'].mean()

# Display the last 5 entries of the resulting Pandas Series
yearly_means.tail()
```

```python
# Group by Month and find the average Mean temperature
result_q9 = df.groupby('Month')['Mean'].mean()
result_q9
```

```python
# Create pivot table with years as rows and months as columns using df_last20
pivot_table_q10 = df_last20.pivot_table(
    values='Mean',
    index='Year',
    columns='Month',
    aggfunc='mean',
    fill_value=0
)
pivot_table_q10
```

```python
import numpy as np
from sklearn.datasets import fetch_openml

print("Loading MNIST dataset from OpenML... (This may take 1-2 minutes)")

# 1. Download the MNIST dataset (70,000 images, each 28x28 pixels)
# as_frame=False ensures data is downloaded as a fast NumPy array
mnist = fetch_openml('mnist_784', version=1, as_frame=False, parser='auto')

# 2. Extract the features into the variable 'X'
# It automatically flattens each 28x28 image into a 1D row of 784 pixels
X = mnist.data

# 3. Normalize pixel values from 0-255 down to a 0.0-1.0 range
# This step is critical for distance-based models like K-Means to perform well
X = X / 255.0

print("Setup Complete!")
print("Features shape (X):", X.shape)  # Expected Output: (70000, 784)
```

```python
from sklearn.cluster import KMeans

# Initialize the K-Means model with 10 clusters
kmeans = KMeans(n_clusters=10, random_state=42)

# Fit the model on the MNIST features X
kmeans.fit(X)

# Assign the cluster labels to the variable kmeans_labels
kmeans_labels = kmeans.labels_

# Display the first few labels
print("Cluster labels:", kmeans_labels[:10])
```

```python
import numpy as np
from sklearn.metrics import f1_score
from scipy.stats import mode

# Create an array to hold the mapped labels
mapped_labels = np.zeros_like(kmeans_labels)

# Map each cluster label to the most frequent true label in that cluster
for i in range(10):
    mask = (kmeans_labels == i)
    if np.any(mask):
        # Find the most frequent true label (mode) in cluster i
        frequent_label = mode(y[mask], keepdims=True).mode[0]
        mapped_labels[mask] = frequent_label

# Calculate the macro-averaged F1 score
f1_macro = f1_score(y, mapped_labels, average='macro')

# Print the final F1 score
print(f"Macro-averaged F1 score: {f1_macro:.4f}")
```
